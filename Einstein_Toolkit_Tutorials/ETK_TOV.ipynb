```python
small_grid = UniformGrid([100, 100], x0=[0, 0], x1=[10,10])

rho_small = rho.read_on_grid(409600, small_grid)

#cf = plt.contourf(*rho_small.coordinates_meshgrid(), np.log10(rho_small.data_xyz), vmin=-10, vmax=-3)
cf = plt.pcolormesh(*rho_small.coordinates_meshgrid(), np.log10(rho_small.data_xyz))
plt.colorbar(cf)
```

```python
%matplotlib notebook
import matplotlib.animation as animation
fig, ax = plt.subplots()
ims = []
for i in rho.iterations:
    rho_small = rho.read_on_grid(i, small_grid)
    im = ax.pcolormesh(*rho_small.coordinates_meshgrid(), np.log10(rho_small.data_xyz), animated=True)
    if i==0:
        ax.pcolormesh(*rho_small.coordinates_meshgrid(), np.log10(rho_small.data_xyz))
    ims.append([im])
    
    
ani = animation.ArtistAnimation(fig, ims, interval=50, blit=True, repeat_delay=1000)
plt.show()
```
